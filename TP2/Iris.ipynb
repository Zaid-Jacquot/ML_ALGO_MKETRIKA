```python
#x = iris["sepal.width"] + iris["petal.width"]
#y = iris["sepal.length"] + iris["petal.length"]

x = iris.loc[:, ["sepal.width"]].values + iris.loc[:, ["petal.width"]].values
y = iris.loc[:, ["sepal.length"]].values + iris.loc[:, ["petal.length"]].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```
